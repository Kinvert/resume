```python
# Setup
%pylab inline
import matplotlib
import numpy as np
from PIL import Image
```

```python
# Open Image
filename = 'sobel.jpg'
im = Image.open(filename)
matplotlib.pyplot.imshow(im) # Check that image loaded properly
```

```python
# Image Prep
im = np.array(im) # Change to Numpy Array
im = im[:,:,0] # From RGB to Grayscale
im = np.invert(im) # Invert Colors
matplotlib.pyplot.imshow(im) # Make sure numpy inverted colors
```

## Smooth Algorithm 1

```python
# Smooth Algorithm 1
w, h = im.shape # Get Width and Height
imtot = np.zeros((w,h)) # Initialize New Image
for i in range(w-1): # Smooth Algoritm
    for j in range(h-1):
        if i > 0 and i < w-1 and j > 0 and j < h-1:
            imtot[i][j] =  ( im[i-1][j-1]*1 + im[i][j-1]*2 + im[i+1][j-1]*1 + \
                            im[i-1][j]*2 + im[i][j]*4 + im[i+1][j]*2 + \
                            im[i-1][j+1]*1 + im[i][j+1]*2 + im[i+1][j+1]*1 ) / 16.0
            
matplotlib.pyplot.imshow(imtot) # Show Results of Smooth
```

## Smooth Algorithm 2

```python
# Smooth Algorithm 2
size = 9
kernel = np.ones((size,size)) # Smooth Kernel
imconv = np.zeros((w,h))
for a in range(w-size+1):
    for b in range(h-size+1):
        imconv[a][b] = np.sum( im[a:a+size, b:b+size]*kernel ) / (size**2)
matplotlib.pyplot.imshow(imconv) # Show Results of Smooth
```
